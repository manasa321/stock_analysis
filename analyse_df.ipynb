```python
import yfinance as yf

def get_financial_metrics(ticker):
    # Fetch data for the ticker
    stock = yf.Ticker(ticker)
    
    # Fetch financial info
    info = stock.info
    
    # Extract required financial metrics
    return {
        'P/E Ratio (TTM)': info.get('trailingPE', None),
        'ROE': info.get('returnOnEquity', None),
        'Debt-to-Equity': info.get('debtToEquity', None),
        'Price-to-Book Ratio': info.get('priceToBook', None),
    }

def create_financial_dataframe(symbols):
    # Create an empty DataFrame with the desired columns
    df = pd.DataFrame(columns=['Symbol', 'P/E Ratio (TTM)', 'ROE', 'Debt-to-Equity', 'Price-to-Book Ratio'])
    
    # List to hold rows as dictionaries
    rows = []
    
    # Fetch and prepare financial metrics for each symbol
    for symbol in symbols:
        metrics = get_financial_metrics(symbol)
        metrics['Symbol'] = symbol
        rows.append(metrics)
        print(metrics)
    
    # Convert list of dictionaries to DataFrame
    df = pd.DataFrame(rows, columns=['Symbol', 'P/E Ratio (TTM)', 'ROE', 'Debt-to-Equity', 'Price-to-Book Ratio'])

    
    return df
```
